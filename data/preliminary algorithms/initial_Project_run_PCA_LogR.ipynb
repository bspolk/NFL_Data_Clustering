```python
import numpy as np
import pandas as pd
```

```python
nfl = pd.read_csv('NFL_Combined.csv')
```

```python
nfl_n = nfl.drop(['Unnamed: 0', 'Rk', 'Tm', 'ID'], axis=1)
```

```python
nfl_Y = pd.DataFrame(nfl['W'])
nfl_Y.head()
```

```python
nfl_Y.loc[nfl_Y['W']>8, 'WinningSeason'] = 1
nfl_Y.loc[nfl_Y['W']<=8, 'WinningSeason'] = 0
```

```python
#nfl_Y = nfl_Y.drop(['W'], axis=1)
```

```python
nfl_Y.head()
```

```python
nfl_X = nfl_n.drop(['W'], axis=1)
```

```python
nfl_Y = nfl_Y.dropna()
```

```python
nfl_X = nfl_X.dropna()
```

```python
nfl_Y.shape
```

```python
nfl_X.describe()
```

```python
nfl_X.corr()
```

```python
# Standardizing and splitting the data
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
train_X, test_X, train_Y, test_Y = train_test_split( nfl_X, nfl_Y, test_size=0.2, random_state=0)
```

```python
scaler = MinMaxScaler()

scaler.fit(train_X)

train_X_scaled = scaler.transform(train_X)
test_X_scaled = scaler.transform(test_X)
```

```python
# PCA
```

```python
from sklearn.decomposition import PCA

pca = PCA(.9)
```

```python
train_pca= pca.fit(train_X_scaled)
```

```python
train_pca.n_components_
```

```python
train_pca_trans = pca.transform(train_X_scaled)
test_pca_trans = pca.transform(test_X_scaled)
```

```python
train_pca_trans.shape
```

```python
# Logistic Regression
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logisticRegr = LogisticRegression(solver = 'lbfgs')
```

```python
logisticRegr.fit(train_pca_trans, train_Y['WinningSeason'])
```

```python
yPred = logisticRegr.predict(test_pca_trans)
```

```python
test_Y['YpredLog']=yPred
```

```python
test_Y.head()
```

```python
logisticRegr.score(test_pca_trans, test_Y['WinningSeason'])
```

```python
# Linear Regression
```

```python
from sklearn.linear_model import LinearRegression
```

```python
LinReg=LinearRegression()
```

```python
LinReg.fit(train_pca_trans, train_Y['W'])
```

```python
ypredLin = LinReg.predict(test_pca_trans)
```

```python
LinReg.score(test_pca_trans, test_Y['W'])
```

```python
test_Y['Ypred'] = ypredLin
```
